# 3. Visualising Distribution of the data :

```python
# 1. distplot(),  2. kdeplot(),  3. jointplot() 
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from numpy.random import randn 
%matplotlib inline
```

```python
x = randn(100000)
```

```python
sns.distplot(x, kde = True, hist =False)
plt.show()
```

```python
sns.distplot(x, kde=False, hist =True)
plt.show()
```

```python
sns.distplot(x, kde=True, hist=True)
plt.show()
```

```python
y =randn(1000)
```

```python
sns.distplot(y,kde=True, hist = True)
plt.show()
```

```python
sns.distplot(y, kde=True, hist=True, shade=False)
plt.show()
```

```python
sns.kdeplot(y, shade=False)
plt.show()
```
