# 2. Categorical Data Ploting :

```python
# 1. catplot(),  2. boxplot(),  3. stripplot(),  4. swarmplot() 
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')

tips.head()
```

```python
tips['sex'].value_counts()
```

```python
## catplot()
```

```python
sns.catplot(x = 'day', y= 'total_bill', data = tips)
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data = tips, hue ='size', kind ='swarm')
plt.show()
```

```python
sns.catplot(x='sex', y='total_bill', data = tips)
plt.show()
```

```python
sns.catplot(x='sex', y = 'total_bill', data=tips, kind = 'swarm', hue = 'size')
plt.show()
```

```python
## swarmplot()
```

```python
sns.swarmplot(x='day', y='total_bill', data = tips)
plt.show()
```

```python
# stripplot()
```

```python
sns.stripplot(x='smoker', y = 'total_bill', data=tips)
plt.show()
```

```python
sns.stripplot(x='sex', y='total_bill', data=tips)
plt.show()
```

```python
sns.stripplot(x='size', y='total_bill', data=tips)           # we can infer different from things by visualisation technique:
plt.show()                                                   # here we can see that people come mostly in group of 2, (so table)
```
