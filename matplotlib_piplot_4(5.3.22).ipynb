```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')

df = iris.drop(["species"], axis = 1)

df.head()
```

```python
d = df.iloc[0]
d
```

```python
d.plot(kind = 'pie', figsize = (8,8))
```

```python
d = df.head(4).T

d
```

```python
d.plot(kind='pie', subplots = True, figsize = (25,20))
plt.show()
```

```python
d.plot(kind='pie', subplots = True, figsize = (25,20), autopct = '%.0f', fontsize = 15)
plt.show()
```
